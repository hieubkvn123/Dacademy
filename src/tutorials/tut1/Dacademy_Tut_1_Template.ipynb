```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

plt.style.use('seaborn-darkgrid')
```

```python
data = pd.read_csv('student-por.csv')
# data.info()
fig, axes = plt.subplots(1, 3, figsize=(21, 8))
sns.kdeplot(data=data, x='G1', hue='romantic', ax=axes[0])
sns.kdeplot(data=data, x='G2', hue='romantic', ax=axes[1])
sns.kdeplot(data=data, x='G3', hue='romantic', ax=axes[2])
plt.show()
```
