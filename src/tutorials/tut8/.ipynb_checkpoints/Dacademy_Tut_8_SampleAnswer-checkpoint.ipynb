# Import libraries

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.datasets import make_regression
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
from sklearn.model_selection import train_test_split

### Torch dependencies ###
import torch
import torch.nn as nn
import torch.optim as optim
```

# Import data

```python
# Read in data
data = pd.read_csv('mushrooms.csv')
data.head()
```

```python
# Split targets and features
columns = list(data.columns)
columns.remove('class')

features = data[columns]
targets = data[['class']]

# Since all of the data is categorical, convert them to numeric labels using OrdinalEncoder
features = OrdinalEncoder().fit_transform(features)
targets = OrdinalEncoder().fit_transform(targets)

# Convert targets to one-hot vectors
targets = OneHotEncoder().fit_transform(targets).toarray()

# Scale the features
features = StandardScaler().fit_transform(features)

# Split the dataset to train-test 
X_train, X_test, Y_train, Y_test = train_test_split(features, targets, test_size=0.2)
```

```python
# Convert your numpy arrays to torch tensors
X_train = torch.tensor(X_train)
Y_train = torch.tensor(Y_train)
X_test = torch.tensor(X_test)
Y_test = torch.tensor(Y_test)

# Convert all to float tensors
X_train = X_train.type(torch.FloatTensor)
Y_train = Y_train.type(torch.FloatTensor)
X_test = X_test.type(torch.FloatTensor)
Y_test = Y_test.type(torch.FloatTensor)
```

# Building the network

```python
class SimpleANN(nn.Module):
    def __init__(self, input_dim, n_classes):
        super(SimpleANN, self).__init__()
        self.fc1 = nn.Linear(in_features=input_dim, out_features=16)
        self.norm1 = nn.BatchNorm1d(16)
        self.relu1 = nn.ReLU()
        
        self.fc2 = nn.Linear(in_features=16, out_features=n_classes)
        self.output = nn.Softmax()
        
    def forward(self, inputs):
        outputs = self.fc1(inputs)
        outputs = self.norm1(outputs)
        outputs = self.relu1(outputs)
        
        outputs = self.fc2(outputs)
        outputs = self.output(outputs)
        
        return outputs

model = SimpleANN(input_dim=features.shape[1], n_classes=targets.shape[1])
print(model)
```

# Define training loop

```python
model = SimpleANN(input_dim=features.shape[1], n_classes=targets.shape[1])
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

epochs = 100
losses = []
torch.autograd.set_detect_anomaly(True)

for i in range(epochs):
    optimizer.zero_grad()
    
    outputs = model(X_train)
    loss = criterion(outputs, Y_train)
    
    losses.append(loss.item())
    
    print('Iteration #', i+1, ' Loss = ', loss)
    
    loss.backward()
    optimizer.step()
```

# Evaluate the model

```python
# Calculate the accuracy on the testing set
predictions = model(X_test)
predictions = predictions.detach().numpy()
predictions = np.argmax(predictions, axis=1)

Y_test = Y_test.detach().numpy()
Y_test = np.argmax(Y_test, axis=1)

accuracy = accuracy_score(Y_test, predictions)
print('Overall accuracy on testing set = ', accuracy)
```

```python
# Plot the confusion matrix and the loss plot
conf_matrix = confusion_matrix(Y_test, predictions)

fig, ax = plt.subplots(1, 2, figsize=(16, 8))

ax[0].plot(losses)
sns.heatmap(conf_matrix, ax=ax[1], annot=True)

ax[0].set_title('Model loss throughout training iterations')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Losses')

ax[1].set_title('Model confusion matrix')
ax[1].set_xlabel('Labels')
ax[1].set_ylabel('Predictions')
plt.show()
```
