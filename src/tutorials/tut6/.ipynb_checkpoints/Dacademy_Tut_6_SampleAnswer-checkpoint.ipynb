# Import libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.cluster import DBSCAN, KMeans
from sklearn.decomposition import PCA

plt.style.use('seaborn-darkgrid')
```

# I. Cluster analysis

## 1. Import data

### Read and clean data

```python
data = pd.read_csv('CC GENERAL.csv')
data['CREDIT_LIMIT'] = SimpleImputer(strategy='median').fit_transform(data[['CREDIT_LIMIT']])
data['MINIMUM_PAYMENTS'] = SimpleImputer(strategy='median').fit_transform(data[['MINIMUM_PAYMENTS']])


data.info()
data.head()
```

### Clean outliers

```python
# Define a function to detect outlier
def get_boundaries(col):
    # Get the upper bound and lower bound of the column
    Q1, Q3 = col.quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range

# Get all feature columns
feature_cols = list(data.columns)
feature_cols.remove('CUST_ID')

# Loop through all feature columns
for col in feature_cols:
    lower, upper = get_boundaries(data[col])
    
    # Replace with upper bound OR lower bound if outlier detected
    data[col] = np.where(data[col] > upper, upper, data[col])
    data[col] = np.where(data[col] < lower, lower, data[col])
    
# Plot box plots to make sure no more outliers
fig, ax = plt.subplots(5, 4, figsize=(16, 16))

for i, feature in enumerate(feature_cols):
    row = i // 4
    col = i %  4
            
    ax[row][col].boxplot(data[feature])
    ax[row][col].set_title(feature)
    
plt.show()
```

## 2. Spectral clustering
#### Understand clustering.
- To understand spectral clustering let's take a step back and re-define the meaning behind clustering problem. Imagine that your dataset is a collection of graph nodes, every data point represents a node and hence your whole dataset represents a graph. 
- In this graph, every node (data point) is connected based on how similar they are to each other. This similarity is defined by various distance functions (e.g: Euclidean distance).
- Clustering is hence understood as a graph partitioning problem.

#### Understand spectral clustering.
- Now you know that clustering can be understood as a graph partitioning problem. The reason why we perform **spectral clustering** is because we want to simplify this graph by reduce the dimensionality of the whole dataset. The rationale behind this is because not all dimensions of data points are relevant to the shape of the graph. Dimensionality reduction is often perform by taking the eigen vectors of the similarity matrix (In case of PCA, this similarity matrix is the covariance matrix). Then, clustering algorithms are applied on this reduced dimensionality instead of the original dimensionality.

### 2.1. Perform PCA on the whole dataset

### 2.1.1. Plot the explained variance for each principle components

```python
# Remove CUST_ID from the features
columns = list(data.columns)
columns.remove('CUST_ID')

# Get the features and scale them
features = data[columns]
features = StandardScaler().fit_transform(features)

# Perform PCA on the selected features
pca = PCA(n_components = None)
pca.fit(features)
evr = pca.explained_variance_ratio_

print('Sum of explained variance in the first three components is : ', evr[:3].sum())

# Plot the EVR curve
fig, ax = plt.subplots(figsize=(15, 8))
ax.plot(evr, marker='o')
ax.set_xlabel('Principle component number')
ax.set_ylabel('Explained Variance Ratio (EVR)')

plt.show()
```

### 2.1.2. Take the first 3 principle components

```python
# Perform PCA on the selected features and get 3 principle components
pca = PCA(n_components = 3)
p_components = pca.fit_transform(features)

# Import the principle components back into the dataframe
data['pc_1'] = p_components[:, 0]
data['pc_2'] = p_components[:, 1]
data['pc_3'] = p_components[:, 2]

# Display the data
data.head()
```

### 2.2. Perform clustering with K-Means

### 2.2.1. Determine how many clusters to pick

```python
wcss = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(data[['pc_1', 'pc_2', 'pc_3']])
    
    print(f'[INFO] Fitting for k = {k}, WCSS = {kmeans.inertia_}')
    wcss.append(kmeans.inertia_)
    
    
fig, ax = plt.subplots(figsize=(15, 8))
ax.plot(wcss, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")

plt.show()
```

### 2.2.2. Perform KMeans

```python
# Let's pick k = 6
kmeans = KMeans(n_clusters = 6)

# Fit on the principle components
kmeans.fit(data[['pc_1', 'pc_2', 'pc_3']])

# Assign cluster IDs
data['CLUSTER_ID'] = kmeans.predict(data[['pc_1', 'pc_2', 'pc_3']])
data.head()
```

# II. Extract insights from the clusters

### 1. Count number of instances in each cluster

```python
sns.countplot(data=data, x='CLUSTER_ID',
             palette=['red', 'green', 'blue', 'cyan', 'yellow', 'purple'])
```

### 2. Plot KDE plots for each features conditioned on the clusters

```python
fig, ax = plt.subplots(4, 4, figsize=(21, 16))
for i, feature in enumerate(feature_cols):
    # Skip tenure because it is fixed anyway
    if(feature == 'TENURE') : continue 
    
    row, col = i // 4, i % 4
    
    sns.kdeplot(data=data, x=feature, hue='CLUSTER_ID',
           palette=['red', 'green', 'blue', 'cyan', 'yellow', 'purple'], ax=ax[row][col])
    ax[row][col].set_title(f'KDE plot for {feature}')
    
plt.tight_layout()
plt.show()
```

# Extracted insights
- The most predominant clusters are clusters number 2-3-4. Clusters 2 and 4 has bears the resemblance that the customers in these clusters are typically keeping very little money in their bank account (As seen in kde plot for BALANCE). This is also reflected in the fact that these two clusters are also low in CREDIT_LIMIT, PAYMENTS and MINIMUM_PAYMENTS.
