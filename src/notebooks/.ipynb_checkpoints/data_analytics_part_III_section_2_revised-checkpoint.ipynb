# Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt

# Torch dependencies
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
```

# Create an image DataLoader 

```python
# Create a preprocessing pipeline
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

# Create a dataset
dataset = datasets.ImageFolder('/home/hieu/Downloads/DOG_CAT/train/', transform=transform)

# Create a dataloader
dataloader = torch.utils.data.DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
# Get some info from the data loader
num_batches = len(dataloader)
batch_size = dataloader.batch_size
class_names = dataset.class_to_idx
class_names = {y : x for x, y in class_names.items()}

print(f'[INFO] Number of batches in dataset : {num_batches}')
print(f'[INFO] Batch size : {batch_size}')
print(f'[INFO] Class IDs to Class names : {class_names}')

# Show sample images
images, labels = next(iter(dataloader))
fig, ax = plt.subplots(4, 4, figsize=(16,16))

for i in range(16):
    row = i // 4
    col = i % 4
    
    class_id = int(labels[i].detach().numpy())
    ax[row][col].imshow(images[i].permute(1,2,0))
    ax[row][col].set_title(class_names[class_id])
    
plt.show()
```

# Create a convolutional neural network
#### Let's create a convolutional neural network with activation functions

```python
# Create a block with conv2d - batch_norm - relu
class CNN(nn.Module):
    def __init__(self, in_channels, n_classes, ksize=3):
        super(CNN, self).__init__()
        # First conv block
        self.conv1 = nn.Conv2d(in_channels=in_channels, out_channels=16, kernel_size=ksize)
        self.relu1 = nn.ReLU()
        
        # Second conv block
        self.conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=ksize)
        self.relu2 = nn.ReLU()
        
        # Max-pooling
        self.maxpool = nn.MaxPool2d(kernel_size=2, stride=2)
        
        # Linear and output layers
        self.fc = nn.Linear(in_features=32*126*126, out_features=64)
        self.hidden = nn.Linear(in_features=64, out_features=n_classes)
        self.output = nn.Sigmoid()
        
    def forward(self, inputs):
        # Output size = in_size - kernel_size + 1
        # After this output_dimension = 254x254 
        outputs = self.conv1(inputs)
        outputs = self.relu1(outputs)
        
        # After this output_dimension = 252x252
        outputs = self.conv2(outputs)
        outputs = self.relu2(outputs)
        
        # After this output dimension = 126 x 126 (with 32 channels)
        outputs = self.maxpool(outputs)
        
        outputs = outputs.reshape(outputs.size(0), -1)
        outputs = self.fc(outputs)
        outputs = self.hidden(outputs)
        outputs = self.output(outputs)
        return outputs
    
model = CNN(in_channels=3, n_classes=2, ksize=3)
print(model)
```

#### Define the training loop

```python
# Switch to GPU if availabel
device = torch.device('cpu')
if(torch.cuda.is_available()):
    print('GPU is available')
    device = torch.device('cuda')
else:
    print('GPU is not available')
    
num_batches = len(dataloader) # Number of batches to train in one epoch
epochs = 3

# Prepare the loss and the model and the optimizer
model = CNN(in_channels=3, n_classes=2, ksize=3)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Define an array to store the batch-wise losses
train_losses = []
val_losses = []

for i in range(epochs):
    for j, (images, labels) in enumerate(dataloader):
        optimizer.zero_grad()
        images = images.to(device)
        labels = labels.to(device)
        
        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)
        train_losses.append(loss.item())
        
        # Backward and optimize
        loss.backward()
        optimizer.step()
        
        if (j+1) % 50 == 0:
            print ('Epoch [{}/{}], Step [{}/{}], Loss: {:.4f}' 
                   .format(i+1, epochs, j+1, num_batches, loss.item()))
```
