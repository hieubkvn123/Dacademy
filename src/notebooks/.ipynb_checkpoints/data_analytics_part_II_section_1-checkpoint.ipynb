# How simple regression works
#### Simple regression formula
![Simple Regression Illustration](https://www.gstatic.com/education/formulas2/355397047/en/simple_linear_regression.svg)
In simple regression, we deduce a value (namely dependent variable) based on a set of features or attributes (namely independent variables) by looking for a linear equation that maps the independent to the dependent variables with minimum error. In order to do that, we have optimize a set of parameters betas using gradient descent.

#### MSE - measure of error in linear regression
![MSE formula](https://cdn-media-1.freecodecamp.org/images/hmZydSW9YegiMVPWq2JBpOpai3CejzQpGkNG)
MSE is a loss function, or in other words, a measure of how well or how badly a model performs. The task of a data analyst while doing regression analysis is the find the parameters set (beta) that minimizes this loss function

#### Gradient Descent - algorithm for minimizing loss
![Gradient Descent Illustration](https://miro.medium.com/max/768/1*CNb8ug5b44Q2CjAhXVUzBw.gif)

# Loading and preprocessing data

```python
# Loading dataset
data = pd.read_csv('../datasets/housing.csv')

# Create a fully integrated preprocessing pipeline
cat_columns = ['ocean_proximity']
num_columns = list(data.select_dtypes('float64').columns)

# Remove class column from numeric columns
num_columns.remove('median_house_value')

pipeline = Pipeline([
    ('imputation', SimpleImputer(strategy='median')),
    ('standard_scaler', StandardScaler())
])

column_transformer = ColumnTransformer([
    ('num', pipeline, num_columns),
    ('cat', OrdinalEncoder(), cat_columns)
])

# Preprocess data and convert the numpy array to pandas dataframe
preprocessed_data = column_transformer.fit_transform(data)
features = pd.DataFrame(data=preprocessed_data, columns=num_columns+cat_columns)
targets = data['median_house_value']

features.head()
```

# Linear Regression model revision

```python
cross_val_losses = []

### Performing cross-validation manually ###
strata_split = StratifiedShuffleSplit(n_splits=10)
stratas = pd.cut(targets, bins=5, labels=[1,2,3,4,5])
for train_index, val_index in strata_split.split(features, stratas):
    X_train = features.loc[train_index]
    Y_train = targets.loc[train_index]
    
    X_val = features.loc[val_index]
    Y_val = targets.loc[val_index]
    
    model = LinearRegression()
    model.fit(X_train, Y_train)
    
    predictions = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(Y_val, predictions))
    cross_val_losses.append(rmse)
    
    print('Validation done, RMSE in this round is ', rmse)
    
print(' --> Mean RMSE : ', np.array(cross_val_losses).mean())
```

# Can we approach regression another way?
* Actually, we can. Sometimes, predicting exact values in regression problems is way too hard to accomplish. Therefore, instead of predicting house price in its exact value, we can divide house prices into ranges. 

* So now instead of predicting the exact house price value, given the inputs, we want to predict in what range the house price will fall into.

```python
num_bins = 5 # number of ranges we want to cut
min_val = targets.min()
max_val = targets.max()

print('The minimum value of house price is ', min_val)
print('The maximum value of house price is ', max_val)

interval = (max_val - min_val)/num_bins
bins = [min_val + i * interval for i in range(num_bins+1)]
print('The bin ranges are : ', bins)

pd.cut(targets, bins=num_bins, labels=[1,2,3,4,5]).hist()
plt.show()
```

```python
# assigns these 'classes' to the target columns
targets = np.asarray(pd.cut(targets, bins=5, labels=[1,2,3,4,5]))
```

# How to exactly classify data into classes ? 
## 1. Option 1 - Logistic Regression
![Logistic Regression formula](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRkmUqpUMx4RFppHTRAl8P67POfFSJEWp2Qj_tQdIFW27HLom8eI2kJJHwh_1r7CPjbzQ&usqp=CAU)
* Logistic regression separate two classes of data using a sigmoid function as stated above. Different from linear regression, the linear function serves as the power of the natural log base in the denominator. However, the task is pretty much the same, finding the parameters set that minimize the log-loss function.
![Log loss formula](https://miro.medium.com/max/1192/1*wilGXrItaMAJmZNl6RJq9Q.png)

### 1.1. Building the model

```python
model = LogisticRegression(max_iter=2000)
# model = DecisionTreeClassifier()

def display_evaluation(model, features, targets):
    # Train-test splitting
    X_train, X_test, Y_train, Y_test = train_test_split(features, targets, test_size=0.333)
    model.fit(features, targets)

    predictions = model.predict(X_test)
    # To evaluate a classification method, we use different metrics - accuracy, f1 score, precision, recall, ...
    # precision = #true_positives / #positive_predictions
    # recall    = #true_positives / #positive_observations
    # f1        = 2 * (precision * recall) /  (precision + recall) => a ballance of precision and recall
    f1 = f1_score(Y_test, predictions, average='micro')
    precision = precision_score(Y_test, predictions, average='micro')
    recall = recall_score(Y_test, predictions, average='micro')
    accuracy = accuracy_score(Y_test, predictions)

    print('Accuracy score of the model : ', accuracy)
    print('Precision score of the model : ', precision)
    print('Recall score of the model : ', recall)
    print('F1 score of the model : ', f1)
    
display_evaluation(model, features, targets)
```

### 1.2. Plotting confusion matrix

```python
plot_confusion_matrix(model, features, targets)
plt.show()
```

## 2. Option 2, more robust - Decision Tree
![Decision Tree example](https://academy.pega.com/sites/default/files/styles/1024/public/media/images/2020-03/decision-tree-concept_2.png?itok=y8hnhcas)

* Decision tree is basically a set of hierachical conditions that we use to separate data into groups. 
* Inorder to decide which condition the tree will split by, decision tree use a measure of "impurity". Measure of impurity basically indicates how "mixed up" the samples are after splitting. 
![Impurity illustration](https://i.ytimg.com/vi/6qDCGI3l-Oo/maxresdefault.jpg)
* The most commonly and simple impurity measure is called the Gini index.
![Gini Index formula](https://media.geeksforgeeks.org/wp-content/uploads/20200620175232/Screenshot-2020-06-20-at-5.51.42-PM.png)
* More information on decision tree, checkout this youtube link: [Youtube](https://www.youtube.com/watch?v=7VeUPuFGJHk)

### 2.1. Build a decision tree

```python
model = DecisionTreeClassifier(max_depth=25)
display_evaluation(model, features, targets)
```

```python
### Visualize the first three levels in the decision trees
fig, ax = plt.subplots(figsize=(15,10))
plot_tree(model, fontsize=10, max_depth=3)
plt.show()
```

### 2.2. Confusion matrix

```python
plot_confusion_matrix(model, features, targets)
plt.show()
```

## 3. Other options - Ensembling methods
#### Example of using random forest - use of many "weak" decision trees

```python
model = RandomForestClassifier(n_estimators = 50) # n_estimators = number of decision trees used
display_evaluation(model, features, targets)
```
