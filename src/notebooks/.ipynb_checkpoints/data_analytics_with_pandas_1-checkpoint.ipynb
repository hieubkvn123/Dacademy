# Grouping
Grouping operation is used to combine initially splitted big group of data and perform reduction operations (averaging, summing, ...) in order to extract a more generalized insight into those big groups.

#### Example 1 : grouping data by weather forecast summary and see the average humidity and temperature of these summaries

```python
'''
    Explain to people that once you do groupby, the column which was grouped by will become
    the index of the resultant dataframe
'''

group_data = data.groupby('Summary') # get the groupby data frame
group_data = group_data.mean() # reduce the group by getting the mean of each group

# sort by humidity and temperature 
mean_humid_data = group_data.sort_values('Humidity')
mean_temp_data = group_data.sort_values('Temperature (C)')

# plot the average humidity and temperature in ascending order
fig, ax = plt.subplots(2,1, figsize=(15, 15))
ax[0].bar(mean_humid_data.index, mean_humid_data['Humidity'].values)
ax[0].set_xticklabels(mean_humid_data.index, rotation=90)
ax[0].set_title('Mean humidity by summary')

ax[1].bar(mean_temp_data.index, mean_temp_data['Temperature (C)'])
ax[1].set_xticklabels(mean_temp_data.index, rotation=90)
ax[1].set_title('Mean temperature by summary')

plt.tight_layout()
plt.show()
```

#### Example 2 : Grouping data by precipitation type and see temperature with humidity by precipitation types

```python
group_data = data.groupby('Precip Type')[['Humidity', 'Temperature (C)']]
group_data = group_data.mean()

mean_humid_data = group_data.sort_values('Humidity')
mean_temp_data = group_data.sort_values('Temperature (C)')

fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].barh(mean_humid_data.index, mean_humid_data['Humidity'], align='center')
ax[0].set_title('Mean humidity by precipitation')

ax[1].barh(mean_temp_data.index, mean_temp_data['Temperature (C)'])
ax[1].set_title('Mean temperature by precipitation')

plt.show()
```

# Describing your data - Normality test

#### Normality test using histogram and skewness score

```python
columns =  ['Wind Speed (km/h)', 'Humidity', 'Temperature (C)']
column = columns[2]
epsilon = 0.1 # Threshold to deem this dataset normal

### More on these descriptive coefficients later ###
mean = data[column].mean()
stddev = data[column].std()
median = data[column].median()
min_ = data[column].min()
max_ = data[column].max()

print(f'The mean of column {column} is {mean}')
print(f'The median of column {column} is {median}')
print(f'The standard deviation of column {column} is {stddev}')
print(f'The min value of column {column} is {min_}')
print(f'The max value of column {column} is {max_}')

### Calculate the skewness of data (moment-based method) ###
'''
N = len(data[column])
m2 = (1/N*(data[column].values - mean)**2).sum()
m3 = (1/N*(data[column].values - mean)**3).sum()
skewness = (m3/(m2 ** 1.5)) * (np.sqrt(N*(N-1))/(N-2))
'''

N = len(data[column])
m2 = stats.moment(data[column], moment=2)
m3 = stats.moment(data[column], moment=3)
skewness = (m3/(m2**1.5)) * (np.sqrt(N*(N-1))/(N-2))

if(abs(skewness) < epsilon):
    print(f' => Data at column {column} is relatively normally distributed')
elif(skewness > 0):
    print(f' => Data at column {column} is leftward-skewed')
elif(skewness < 0):
    print(f' => Data at column {column} is rightward-skewed')

### Plotting the distribution ###
x_axis = np.arange(min_, max_, 0.001)
ax = data[column].plot.hist(bins=30, alpha=0.5, figsize=(10,5))

### Plot a matching gaussian curve ###
ax2 = ax.twinx()
ax2.plot(x_axis, norm.pdf(x_axis, mean, stddev), color='red')
plt.show()
```

#### Normality test using Shapiro-Wilk

```python
#### Normality test using Shapiro-Wilk method ###
''' 
    Shapiro-Wilk is a powerful method for testing for data normality when
    the sample size is small (n < 50)
    
    H0 : The dataset is in normal distribution
    H1 : The dataset is not in normal distribution
'''
alpha = 0.05
test = stats.shapiro(data[column].sample(50))
W = test.statistic
p = test.pvalue

print(f'The test statistic is {W} and the p value is {p}')
if(p > alpha): # do not reject the null hypothesis
    print(f'Data from column {column} is from a normal distribution')
else: # reject the null hypothesis
    print(f'Data from column {column} is not from a normal distribution')
```
