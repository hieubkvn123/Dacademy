## 1. What is a regression problem?
* A regression problem is a task whereby a set of features or variables is provided in order help predict a set of targeted values. For example, 
    - 1. Based on the historical data of a time series (e.g : stock price, covid cases, ...). Predict the next values in the time series.
    - 2. Based on the financial figures of a companies (ROI, EBITDA, ...). Predict the net income of the company in the next work year.
    - 3. Based on the age, sleep hour and social/romantic relationship status, predict the major depressive index (MDI) of a college student.
    
<br/>
    
* **Common methods for regression problems**
    - **Linear regression (Ordinary Least Squares)**
    - **Decision Tree**
    - Deep learning (RNN, LSTM, GRU, ...)
    - Gaussian Processes
    - Moving Average
    - ...

## 2. Linear regression (OLS)
* Linear regression is a naive approach for regression problems. This algorithm predicts a targeted values by simply finding the best-fit line that aligns with the targeted values.

<br/>

### 2.1. Simple linear regression
* Simple linear regression refers to the linear regression algorithm that uses only one explanatory variable to predict the targeted values. 
* General form of simple linear regression :
## $\hat Y = \alpha + \beta x$

<br/>

* Solution of simple linear regression :
#### $\beta = \frac{\sum (x_{i} - \bar x)(y_{i} - \bar y)}{\sum (x_{i} - \bar x)^2}$
#### $\alpha = \bar y - \beta \bar x $

```python
# Example of simple linear regression
X, Y = make_regression(n_features=1, noise=10)
lr = LinearRegression().fit(X, Y)
print(X.shape, Y.shape)
alpha, beta = lr.intercept_, lr.coef_

Y_hat = lr.predict(X)
X_test = np.linspace(min(X), max(X)).reshape(-1, 1)

fig, ax = plt.subplots(1, 2, figsize=(15, 6))
ax[0].scatter(X, Y)
ax[1].scatter(X, Y)
ax[1].plot(X_test, lr.predict(X_test), color='red')

ax[1].set_xlabel("X")
ax[1].set_ylabel("Y")
plt.show()
```

### 2.2. Multivariable regression
* Unlike simple regression, multivariable regression is used to predict targeted values based on two or more explanatory variables. So the fitted function will be a hyperplane instead of a line in a 2D space.

<br/>
* General form of multivariable regression :

### $ \hat Y = \beta_{0} + \sum \limits _{i=1} ^{P} \beta_{i}x_{i}$

<br/>

* Solution for multivariable linear regression
$ \beta = (X^TX)^-1(X^Ty) $

```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

df = sns.load_dataset('mpg')
df.dropna(inplace=True)

model = smf.ols(formula='mpg ~ horsepower + acceleration', data=df)
results = model.fit()

x, y = model.exog_names[1:]

x_range = np.arange(df[x].min(), df[x].max())
y_range = np.arange(df[y].min(), df[y].max())

X, Y = np.meshgrid(x_range, y_range)

exog = pd.DataFrame({x: X.ravel(), y: Y.ravel()})
Z = results.predict(exog = exog).values.reshape(X.shape)

fig = plt.figure(figsize=plt.figaspect(1)*2)
ax = plt.axes(projection='3d')
ax.scatter(df[x].values, df[y].values, results.fittedvalues.values, 
           marker='.', label="Fits")
cond = df[model.endog_names].values > results.fittedvalues.values
ax.scatter(df[x][cond].values, df[y][cond].values, df[model.endog_names]
           [cond].values, label="Raw")
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, alpha = 0.4)
ax.scatter(df[x][cond == False].values, df[y][cond == False].values,
           df[model.endog_names][cond == False].values)
ax.legend()
plt.show()
```

## 3. Assumptions of OLS
* 1. Linearity of parameters : The parameters used in the regression function must be linear (No quadratic, cubic, ...)
* 2. Zero conditional mean of error : The average error will eventually be zero.
* 3. Random sampling : Our data must have been randomly sampled from the population.
* 4. No perfect correlation : The regressors must not be perfectly correlated (Pearson R = 1 or -1)
* 5. Homoscedasticity : No matter the values of our regressors, the variance of the errors must be constant.
